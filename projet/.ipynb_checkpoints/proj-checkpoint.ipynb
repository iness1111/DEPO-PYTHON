```python
# importation et exploitation du fichier discours_US.csv

import csv
import pandas as pd

# 1.1 Load the CSV
data = []
with open("discours_US.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f, delimiter='\t', quotechar='"')
    for row in reader:
        if len(row) == 5: 
            data.append(row)

# Build DataFrame
df = pd.DataFrame(data, columns=["auteur", "texte", "date", "titre", "url"])

# Preview
print(df.head())
print(df.info())


```

```python
# verif author distribution
print(df["auteur"].value_counts())  
print( df.shape)  
```

```python
# creation et save d un corpus avec le fichier discours_US.csv

from nltk.tokenize import sent_tokenize

# 1.3 Create Corpus
corpus = Corpus(nom="Discours_US")

# Add documents (split speeches into sentences)

# Loop through each row in the CSV
for idx, row in df.iterrows():
    auteur = row["auteur"]
    titre = row["titre"]
    texte = row["texte"]
    date = row["date"]
    url = row["url"]

    # Split text into sentences
    sentences = sent_tokenize(texte)

    # Add each sentence as a Document in the Corpus
    for sent in sentences:
        doc = Corpus.factory(
            source="csv",       
            titre=titre,
            auteur=auteur,
            date=date,
            url=url,
            texte=sent
        )
        corpus.add_document(doc)

print("Corpus size:", corpus.ndoc)
```

```python
corpus.save("DiscoursCorpus.csv")
print("DONE:", corpus)
```
